```python
def lat_convert(x):
    x = float(x[1:])
    return x

lat_convert('(33.9599')
```

```python
df['lat'] = df.lat.apply(lat_convert)
df.head()
```

```python
def lon_convert(x):
    x = float(x[:-1])
    return x

lon_convert('33.9599)')
```

```python
df['lon'] = df.lon.apply(lon_convert)
df.head()
```

Descent Code: A - Other Asian B - Black C - Chinese D - Cambodian F - Filipino G - Guamanian H - Hispanic/Latin/Mexican I - American Indian/Alaskan Native J - Japanese K - Korean L - Laotian O - Other P - Pacific Islander S - Samoan U - Hawaiian V - Vietnamese W - White X - Unknown Z - Asian Indian

```python
def define_descent(x):
    if x == 'W':
        x = 'White'
    elif x == 'H':
        x = 'Hispanic/Latin/Mexican'
    elif x == 'O':
        x = 'Other'
    elif x == 'B':
        x = 'Black'
    elif x == 'A':
        x = 'Other Asian'
    elif x == 'X':
        x = 'Unknown'
    elif x == 'K':
        x = 'Korean'
    elif x == 'L':
        x = 'Loation'
    elif x == 'P':
        x = 'Pacific Islander'
    elif x == 'S':
        x = 'Samoan'
    elif x == 'V':
        x = 'Viatnamese'
    elif x == 'J':
        x = 'Japanese'
    elif x == 'F':
        x = 'Filipino'
    elif x == 'C':
        x = 'Chinese'
    elif x == 'I':
        x = 'American Indian/Alaskan Native'
    elif x == 'U':
        x = 'Hawaiian'
    elif x == 'D':
        x = 'Cambodian'
    elif x == 'G':
        x = 'Guamanian'
    elif x == 'Z':
        x = 'Asian Indian'
    return x
```

```python
df['Victim Descent'] = df['Victim Descent'].apply(define_descent)
```
